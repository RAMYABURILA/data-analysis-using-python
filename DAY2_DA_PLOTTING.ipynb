```python
import numpy as np
from matplotlib import pyplot as plt
ind=[25,51,84,131,160,189,220,250,267,297]
aus=[15,41,94,110,151]
overs=np.arange(1,51,5)
plt.title("IND VS AUS")
plt.plot(overs,ind,label="India",marker=">")
over2=np.arange(1,25,5)
plt.plot(over2,aus,label="Australia",marker="P",color="yellow")
plt.legend()
plt.grid()
```
