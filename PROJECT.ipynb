```python
from sklearn.ensemble import RandomForestClassifier
LR=RandomForestClassifier(n_estimators=10)
```

```python
import pandas as pd
```

```python
file=pd.read_csv("/content/WA_Fn-UseC_-HR-Employee-Attrition.csv")
file
```

```python
feature_cols=['Age','JobSatisfaction']
features=file[feature_cols]
features

```

```python
target=file.Attrition
```

```python
LR.fit(features,target)
```

```python
LR.predict([[41,4]])
```

```python
90
Age=int(input("enter age:"))
sats=int(input("enter job satisfaction rate:"))
pred=LR.predict([[Age,sats]])
if pred=='Yes':
  print("Attrition")
else:
  print("No attrition")
```

```python
#saving all DEcision Trees
import os
output_dir="tree_visualizations"
os.makedirs(output_dir, exist_ok=True)
```

```python
for i, tree in enumerate(LR.estimators_):
  tree_dot_file=os.path.join(output_dir, f"tree_{i}.dot")
  tree_png_file=os.path.join(output_dir, f"tree_{i}.dot")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=["Age", "sast"],
                    class_names=[str(cls) for cls in LR.classes_], filled=True, rounded=True)
```

```python
command=f"dot-Tpng{tree_dot_file}-0{tree_png_file}"
os.system(command)
print(f"Tree {i}visualization saved to{tree_png_file}")
```
